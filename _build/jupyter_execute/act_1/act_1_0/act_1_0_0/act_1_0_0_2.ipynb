   - Hypothesis
   - Cognitive frailty predicts delirium in liver transplant
   - Its defined as: 
    - Cognitive frailty is defined as the presence of both physical frailty and cognitive impairment without dementia
    - Physical frailty is defined as the presence of three or more of the following five criteria: unintentional weight loss, self-reported exhaustion, weakness, slow walking speed, and low physical activity
    - Cognitive impairment without dementia is defined as a score of 0.5 standard deviations or more below the mean on two or more tests of cognitive function
    - Cognitive frailty is a risk factor for delirium in liver transplant recipients
